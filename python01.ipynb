```python
# Simple Calculator in Python

num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /): ")
num2 = float(input("Enter second number: "))

if operator == "+":
    print("Result:", num1 + num2)
elif operator == "-":
    print("Result:", num1 - num2)
elif operator == "*":
    print("Result:", num1 * num2)
elif operator == "/":
    if num2 == 0:
        print("Error: Cannot divide by zero!")
    else:
        print("Result:", num1 / num2)
else:
    print("Invalid operator!")
```

```python
import os
fn = "marks.txt"
if not os.path.exists(fn):
    with open(fn, "w") as f:
        f.write("80\n90\n75\n60\n95\n")
    print(f"'{fn}' not found — created a sample '{fn}'. (You can upload your own file in Colab sidebar.)")

marks = []
with open(fn) as f:
    for i, line in enumerate(f, 1):
        line = line.strip()
        if not line:
            continue
        try:
            marks.append(float(line))
        except ValueError:
            print(f"Skipping invalid line {i}: {repr(line)}")

if not marks:
    result = "No valid marks found."
else:
    avg = sum(marks) / len(marks)
    result = f"Average Marks = {avg:.2f}  (from {len(marks)} entries)"

with open("result.txt", "w") as out:
    out.write(result)

print("Done. Check 'result.txt' —", result)
```

```python
import re

emails = ["diya@gmail.com", "rawat@@123", "hello@yahoo.com", "abc@", "test@domain.org"]

pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'

valid = [e for e in emails if re.match(pattern, e)]
invalid = [e for e in emails if e not in valid]

print("Valid:", valid)
print("Invalid:", invalid)
```

```python
import datetime

employees = {
    "Diya": "09:45",
    "Rohan": "09:20",
    "Neha": "10:05",
    "Aman": "09:10"
}

late = []

for name, time in employees.items():
    if datetime.datetime.strptime(time, "%H:%M") > datetime.datetime.strptime("09:30", "%H:%M"):
        late.append(name)

print("Late Employees:", late)
```

```python
from multiprocessing import Pool

def square(n):
    return n*n

if __name__ == "__main__":
    nums = [10000, 20000, 30000, 40000, 50000]

    with Pool() as p:
        result = p.map(square, nums)

    print("Squares:", result)
```

```python
import threading, urllib.request

# URLs of small files for testing
urls = [
    "https://raw.githubusercontent.com/github/gitignore/main/Python.gitignore",
    "https://raw.githubusercontent.com/github/gitignore/main/Java.gitignore",
    "https://raw.githubusercontent.com/github/gitignore/main/C++.gitignore"
]

def download_file(url):
    filename = url.split("/")[-1]
    urllib.request.urlretrieve(url, filename)
    print(f"Downloaded: {filename}")

threads = []
for url in urls:
    t = threading.Thread(target=download_file, args=(url,))
    t.start()
    threads.append(t)

for t in threads:
    t.join()

print("All downloads completed.")
```
